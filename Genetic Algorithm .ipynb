```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
```

```python
#defining the function to be minimized
def f(x):
    return x[0]*x[1]-x[2]    #it will take an array as input and return the sum of the array

    
```

```python
varbound=np.array([[2,6],[3,8],[3,5]])
vartype=np.array([['real'],['int'],['int']])

model=ga(function=f,dimension=3,variable_type_mixed=vartype,variable_boundaries=varbound)

model.run()
```

```python
#setting up the arguments in GA
algorithm_param = {'max_num_iteration': 3000,\
                   'population_size':500,\
                   'mutation_probability':0.1,\
                   'elit_ratio': 0.01,\
                   'crossover_probability': 0.8,\
                   'parents_portion': 0.3,\
                   'crossover_type':'uniform',\
                   'max_iteration_without_improv':None}
model=ga(function=f,dimension=3,variable_type_mixed=vartype,variable_boundaries=varbound,algorithm_parameters=algorithm_param)

model.run()
```
